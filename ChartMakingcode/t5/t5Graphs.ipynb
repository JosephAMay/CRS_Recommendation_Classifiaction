```python
#Load in Results
import numpy as np
import torch
import lightning.pytorch as pl
from torchinfo import summary
from torchview import draw_graph
import matplotlib.pyplot as plt
import pandas as pd


t5_NONstandardized_validation_accuracy = np.loadtxt("results_t5_NONstanadrdized_validation_accuracy.txt")
t5_NONstandardized_validation_loss = np.loadtxt("results_t5_NONstanadrdized_validation_loss.txt")

t5_standardized_validation_loss = np.loadtxt("results_t5_standardized_validation_loss.txt")
t5_standardized_validation_accuracy = np.loadtxt("results_t5_standardized_validation_accuracy.txt")
#t5 Training
t5_NONstandardized_train_accuracy = np.loadtxt("results_t5_NONstanadrdized_train_accuracy.txt")  
t5_NONstandardized_train_loss = np.loadtxt("results_t5_NONstanadrdized_train_loss.txt")  

t5_standardized_train_accuracy = np.loadtxt("results_t5_standardized_train_accuracy.txt")  
t5_standardized_train_loss = np.loadtxt("results_t5_standardized_train_loss.txt")  
```

# t5 Validation Results

```python

plt.plot(np.arange(0,t5_standardized_validation_accuracy.shape[1]),
 np.mean(t5_standardized_validation_accuracy,0),label="Accuracy")
plt.fill_between(np.arange(0,t5_standardized_validation_accuracy.shape[1]),
     np.mean(t5_standardized_validation_accuracy,0)-(1.96*np.std(t5_standardized_validation_accuracy,0)/np.sqrt(t5_standardized_validation_accuracy.shape[0])),
     np.mean(t5_standardized_validation_accuracy,0)+(1.96*np.std(t5_standardized_validation_accuracy,0)/np.sqrt(t5_standardized_validation_accuracy.shape[0])),
     alpha=0.5)
plt.plot(np.arange(0,t5_standardized_validation_loss.shape[1]),
 np.mean(t5_standardized_validation_loss,0),label="Loss")
plt.fill_between(np.arange(0,t5_standardized_validation_loss.shape[1]),
 np.mean(t5_standardized_validation_loss,0)-(1.96*np.std(t5_standardized_validation_loss,0)/np.sqrt(t5_standardized_validation_loss.shape[0])),
 np.mean(t5_standardized_validation_loss,0)+(1.96*np.std(t5_standardized_validation_loss,0)/np.sqrt(t5_standardized_validation_loss.shape[0])),
 alpha=0.5)
plt.ylim([0.4,1])
plt.ylabel('Validation Accuracy & Loss (P)')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
plt.title('t5-Standardized Validation Accuracy & Loss')
plt.show()
```
