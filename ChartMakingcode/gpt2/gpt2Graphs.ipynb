```python

plt.plot(np.arange(0,gpt2_standardized_validation_accuracy.shape[1]),
 np.mean(gpt2_standardized_validation_accuracy,0),label="Accuracy")
plt.fill_between(np.arange(0,gpt2_standardized_validation_accuracy.shape[1]),
     np.mean(gpt2_standardized_validation_accuracy,0)-(1.96*np.std(gpt2_standardized_validation_accuracy,0)/np.sqrt(gpt2_standardized_validation_accuracy.shape[0])),
     np.mean(gpt2_standardized_validation_accuracy,0)+(1.96*np.std(gpt2_standardized_validation_accuracy,0)/np.sqrt(gpt2_standardized_validation_accuracy.shape[0])),
     alpha=0.5)
plt.plot(np.arange(0,gpt2_standardized_validation_loss.shape[1]),
 np.mean(gpt2_standardized_validation_loss,0),label="Loss")
plt.fill_between(np.arange(0,gpt2_standardized_validation_loss.shape[1]),
 np.mean(gpt2_standardized_validation_loss,0)-(1.96*np.std(gpt2_standardized_validation_loss,0)/np.sqrt(gpt2_standardized_validation_loss.shape[0])),
 np.mean(gpt2_standardized_validation_loss,0)+(1.96*np.std(gpt2_standardized_validation_loss,0)/np.sqrt(gpt2_standardized_validation_loss.shape[0])),
 alpha=0.5)
plt.ylim([0.4,1])
plt.ylabel('Validation Accuracy & Loss (P)')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
plt.title('GPT2-Standardized Validation Accuracy & Loss')
plt.show()
```
