```python
#Load in Results
import numpy as np
import torch
import lightning.pytorch as pl
from torchinfo import summary
from torchview import draw_graph
import matplotlib.pyplot as plt
import pandas as pd


#GPT NEO Validation
neo_NONstandardized_validation_accuracy = np.loadtxt("results_neo_NONstanadrdized_validation_accuracy.txt")
neo_NONstandardized_validation_loss = np.loadtxt("results_neo_NONstanadrdized_validation_loss.txt")

neo_standardized_validation_accuracy = np.loadtxt("results_neo_stanadrdized_validation_accuracy.txt")  
neo_standardized_validation_loss = np.loadtxt("results_neo_stanadrdized_validation_loss.txt")  

#GPT NEO Training
neo_NONstandardized_train_accuracy = np.loadtxt("results_neo_NONstanadrdized_train_accuracy.txt")
neo_NONstandardized_train_loss = np.loadtxt("results_neo_NONstanadrdized_train_loss.txt")

neo_standardized_train_accuracy = np.loadtxt("results_neo_stanadrdized_train_accuracy.txt")  
neo_standardized_train_loss= np.loadtxt("results_neo_stanadrdized_train_loss.txt")
```

# NEO Validation Results

```python

plt.plot(np.arange(0,neo_standardized_validation_accuracy.shape[1]),
 np.mean(neo_standardized_validation_accuracy,0),label="Accuracy")
plt.fill_between(np.arange(0,neo_standardized_validation_accuracy.shape[1]),
     np.mean(neo_standardized_validation_accuracy,0)-(1.96*np.std(neo_standardized_validation_accuracy,0)/np.sqrt(neo_standardized_validation_accuracy.shape[0])),
     np.mean(neo_standardized_validation_accuracy,0)+(1.96*np.std(neo_standardized_validation_accuracy,0)/np.sqrt(neo_standardized_validation_accuracy.shape[0])),
     alpha=0.5)
plt.plot(np.arange(0,neo_standardized_validation_loss.shape[1]),
 np.mean(neo_standardized_validation_loss,0),label="Loss")
plt.fill_between(np.arange(0,neo_standardized_validation_loss.shape[1]),
 np.mean(neo_standardized_validation_loss,0)-(1.96*np.std(neo_standardized_validation_loss,0)/np.sqrt(neo_standardized_validation_loss.shape[0])),
 np.mean(neo_standardized_validation_loss,0)+(1.96*np.std(neo_standardized_validation_loss,0)/np.sqrt(neo_standardized_validation_loss.shape[0])),
 alpha=0.5)
plt.ylim([0.3,1])
plt.ylabel('Validation Accuracy & Loss (P)')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
plt.title('NEO-Standardized Validation Accuracy & Loss')
plt.show()
```

```python
plt.plot(np.arange(0,neo_NONstandardized_validation_accuracy.shape[1]),
 np.mean(neo_NONstandardized_validation_accuracy,0),label="Nonstandardized Accuracy")
plt.fill_between(np.arange(0,neo_NONstandardized_validation_accuracy.shape[1]),
     np.mean(neo_NONstandardized_validation_accuracy,0)-(1.96*np.std(neo_NONstandardized_validation_accuracy,0)/np.sqrt(neo_NONstandardized_validation_accuracy.shape[0])),
     np.mean(neo_NONstandardized_validation_accuracy,0)+(1.96*np.std(neo_NONstandardized_validation_accuracy,0)/np.sqrt(neo_NONstandardized_validation_accuracy.shape[0])),
     alpha=0.5)
plt.plot(np.arange(0,neo_standardized_validation_accuracy.shape[1]),
 np.mean(neo_standardized_validation_accuracy,0),label="Standardized Accuracy")
plt.fill_between(np.arange(0,neo_standardized_validation_accuracy.shape[1]),
 np.mean(neo_standardized_validation_accuracy,0)-(1.96*np.std(neo_standardized_validation_accuracy,0)/np.sqrt(neo_standardized_validation_accuracy.shape[0])),
 np.mean(neo_standardized_validation_accuracy,0)+(1.96*np.std(neo_standardized_validation_accuracy,0)/np.sqrt(neo_standardized_validation_accuracy.shape[0])),
 alpha=0.5)
plt.ylim([0.35,.95])
plt.ylabel('Validation Accuracy (P)')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
plt.title('NEO Standardized vs Nonstandardized Validation Accuracy')
plt.show()
```
