```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def shapStripPlot(modelName):
    # Convert the SHAP values array to a NumPy array
    #shap_array = np.array(myarr)
    shap_array = np.loadtxt(modelName+"_shap.txt")
    # Convert the reshaped array to a pandas DataFrame with appropriate column names
    feature_names = ['Length', 'Readability', 'Word Importance', 'Repetition', 
                     'Polarity', 'Subjectivity', 'Feature Appearance', 'Grammar']
    shap_df = pd.DataFrame(shap_array, columns=feature_names)
    
    # Melt the DataFrame to create a long-form DataFrame suitable for plotting
    shap_melted = shap_df.melt(var_name='Feature', value_name='SHAP Value')
    
    # Create the bee swarm plot using Seaborn
    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 6))
    sns.stripplot(x='Feature', y='SHAP Value', data=shap_melted, size = 4)
    plt.title(modelName.upper()+' SHAP Values for Each Quality Factor',fontsize=18)
    plt.xlabel('Quality Factor',fontsize=18)
    plt.ylabel('SHAP Value',fontsize=18)
    plt.xticks(rotation=45,fontsize=16)  # Rotate x-axis labels for better readability
    plt.tight_layout()
    plt.savefig('shap_'+modelName+'_chart.png')
    plt.show()
```
