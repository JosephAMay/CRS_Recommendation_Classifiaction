# Standardized Validation Results

```python

plt.plot(np.arange(0,gpt2_standardized_validation_accuracy.shape[1]),
 np.mean(gpt2_standardized_validation_accuracy,0),label="GPT2 Accuracy")
plt.fill_between(np.arange(0,gpt2_standardized_validation_accuracy.shape[1]),
     np.mean(gpt2_standardized_validation_accuracy,0)-(1.96*np.std(gpt2_standardized_validation_accuracy,0)/np.sqrt(gpt2_standardized_validation_accuracy.shape[0])),
     np.mean(gpt2_standardized_validation_accuracy,0)+(1.96*np.std(gpt2_standardized_validation_accuracy,0)/np.sqrt(gpt2_standardized_validation_accuracy.shape[0])),
     alpha=0.5)
plt.plot(np.arange(0,neo_standardized_validation_accuracy.shape[1]),
 np.mean(neo_standardized_validation_accuracy,0),label="NEO Accuracy")
plt.fill_between(np.arange(0,neo_standardized_validation_accuracy.shape[1]),
 np.mean(neo_standardized_validation_accuracy,0)-(1.96*np.std(neo_standardized_validation_accuracy,0)/np.sqrt(neo_standardized_validation_accuracy.shape[0])),
 np.mean(neo_standardized_validation_accuracy,0)+(1.96*np.std(neo_standardized_validation_accuracy,0)/np.sqrt(neo_standardized_validation_accuracy.shape[0])),
 alpha=0.5)
plt.plot(np.arange(0,t5_standardized_validation_accuracy.shape[1]),
 np.mean(t5_standardized_validation_accuracy,0),label="t5 Accuracy")
plt.fill_between(np.arange(0,t5_standardized_validation_accuracy.shape[1]),
 np.mean(t5_standardized_validation_accuracy,0)-(1.96*np.std(t5_standardized_validation_accuracy,0)/np.sqrt(t5_standardized_validation_accuracy.shape[0])),
 np.mean(t5_standardized_validation_accuracy,0)+(1.96*np.std(t5_standardized_validation_accuracy,0)/np.sqrt(t5_standardized_validation_accuracy.shape[0])),
 alpha=0.5)
plt.ylim([0.4,1])
plt.ylabel('Validation Accuracy (P)')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
plt.title('GPT2,NEO,t5 Standardized Validation Accuracy')
plt.show()
```
